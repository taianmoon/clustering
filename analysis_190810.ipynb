```python
%matplotlib notebook
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
import math
import glob
import os.path

method_name='MeanShift'  #======================================HERE:KMeans/PCA/MeanShift

hdul = fits.open('output_combined_190810.fits')
data = hdul[1].data

ra_pre = data.field('ra')
dec_pre = data.field('dec')
f250_pre = data.field('f350')
et250_pre = data.field('et250')
f350_pre = data.field('f350')
et350_pre = data.field('et350')
f500_pre = data.field('f500')
et500_pre = data.field('et500')
color_f250_f350_pre = data.field('f250/f350')
color_f350_f500_pre = data.field('f350/f500')

output_kmeans_pre = data.field('output_kmeans')
output_pca_pre = data.field('output_pca')
output_meanshift_pre = data.field('output_meanshift')

#ra=[]
#dec=[]
#f250=[]
#et250=[]
#f350=[]
#et350=[]
#f500=[]
#et500=[]
#color_f250_f350=[]
#color_f350_f500=[]

color_sequence=[]
size_sequence=[]
#shape_sequence=[]
for i in range(0, len(ra_pre)):
    
    if method_name=='KMeans':
        if output_kmeans_pre[i] == 0.0:
            color_sequence.append('blue')
            size_sequence.append(1.0)
            #shape_sequence.append('.')
        elif output_kmeans_pre[i] == 1.0:
            color_sequence.append('red')
            size_sequence.append(10.0)
            #shape_sequence.append('.')
        else:
            color_sequence.append('black')
            size_sequence.append(10.0)
            #shape_sequence.append('.')
            
    elif method_name=='PCA':
        if output_pca_pre[i] == 0.0:
            color_sequence.append('blue')
            size_sequence.append(1.0)
            #shape_sequence.append('.')
        elif output_pca_pre[i] == 1.0:
            color_sequence.append('red')
            size_sequence.append(10.0)
            #shape_sequence.append('.')
        elif output_pca_pre[i] == 2.0:
            color_sequence.append('black')
            size_sequence.append(1.0)
            #shape_sequence.append('.')
        elif output_pca_pre[i] == 3.0:
            color_sequence.append('green')
            size_sequence.append(1.0)
            #shape_sequence.append('.')
        else:
            color_sequence.append('purple')
            size_sequence.append(10.0)
            #shape_sequence.append('.')
    
    elif method_name=='MeanShift':
        if output_meanshift_pre[i] == 0.0:
            color_sequence.append('blue')
            size_sequence.append(1.0)
            #shape_sequence.append('.')
        elif output_meanshift_pre[i] == 1.0:
            color_sequence.append('red')
            size_sequence.append(10.0)
            #shape_sequence.append('.')
        elif output_meanshift_pre[i] == 2.0:
            color_sequence.append('black')
            size_sequence.append(10.0)
            #shape_sequence.append('.')
        elif output_meanshift_pre[i] == 3.0:
            color_sequence.append('green')
            size_sequence.append(10.0)
            #shape_sequence.append('.')
        elif output_meanshift_pre[i] == 4.0:
            color_sequence.append('purple')
            size_sequence.append(10.0)
            #shape_sequence.append('.')
        elif output_meanshift_pre[i] == -1.0:
            color_sequence.append('gray')
            size_sequence.append(1.0)
            #shape_sequence.append('.')
        else:                               #5-10
            color_sequence.append('orange')
            size_sequence.append(1.0)
            #shape_sequence.append('.')



    
#print(type(color_sequence))
#print('gggggggggggggggg')

fig = plt.figure()

plt.scatter(ra_pre, dec_pre, color=color_sequence, s=size_sequence)
plt.xlabel('RA (deg)')
plt.ylabel('Dec (deg)')
plt.grid()
if method_name=='KMeans':
    plt.title(method_name + ', 0:blue; 1:red; 2:black')
elif method_name=='PCA':
    plt.title(method_name + ', 0:blue; 1:red; 2:black; 3:green; 4:purple')
elif method_name=='MeanShift':
    plt.title(method_name + ', -1: grey; 0:blue; 1:red; 2:black; 3:green; 4:purple; 5-10: orange')
    
    
plt.show()


fig = plt.figure()

plt.scatter(color_f250_f350_pre, color_f350_f500_pre, color=color_sequence, s=size_sequence)
plt.xlabel('F250/F350')
plt.ylabel('F350/F500')
#plt.xlim((0.0, 5.0))
#plt.ylim((0.0, 10.0))
plt.grid()
if method_name=='KMeans':
    plt.title(method_name + ', 0:blue; 1:red; 2:black')
elif method_name=='PCA':
    plt.title(method_name + ', 0:blue; 1:red; 2:black; 3:green; 4:purple')
elif method_name=='MeanShift':
    plt.title(method_name + ', -1: grey; 0:blue; 1:red; 2:black; 3:green; 4:purple; 5-10: orange')
 
plt.show()
```
